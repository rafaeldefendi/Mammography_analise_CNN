```python
import os
import shutil

#Images folders directory.
base_dir = r'E:\Coursera\CM3070 - Final Project\dataset\ORGANIZED_final\Mass\Train\BENIGN\ROI'
#base_dir = r'E:\Coursera\CM3070 - Final Project\dataset\ORGANIZED_final\Mass\Train\MALIGNANT\ROI'
#base_dir = r'E:\Coursera\CM3070 - Final Project\dataset\ORGANIZED_final\Mass\Test\BENIGN\ROI'
#base_dir = r'E:\Coursera\CM3070 - Final Project\dataset\ORGANIZED_final\Mass\Test\MALIGNANT\ROI'
#base_dir = r'E:\Coursera\CM3070 - Final Project\dataset\ORGANIZED_final\Calc\Train\BENIGN\ROI'
#base_dir = r'E:\Coursera\CM3070 - Final Project\dataset\ORGANIZED_final\Calc\Train\MALIGNANT\ROI'
#base_dir = r'E:\Coursera\CM3070 - Final Project\dataset\ORGANIZED_final\Calc\Test\BENIGN\ROI'
#base_dir = r'E:\Coursera\CM3070 - Final Project\dataset\ORGANIZED_final\Calc\Test\MALIGNANT\ROI'

for root, dirs, files in os.walk(base_dir):
    for file in files:
        #Search DICOM files.
        if file.lower().endswith('.dcm'):
            source_path = os.path.join(root, file)

            #Case name = first after _ROI.
            relative_path = os.path.relpath(root, base_dir)
            case_name = relative_path.split(os.sep)[0]

            #File name.
            new_filename = f'{case_name}.dcm'
            destination_path = os.path.join(base_dir, new_filename)

            #To avoid overwriting files.
            counter = 1
            while os.path.exists(destination_path):
                destination_path = os.path.join(
                    base_dir, f'{case_name}_{counter}.dcm'
                )
                counter += 1

            #Move and rename.
            shutil.move(source_path, destination_path)
```
