```python
import pandas as pd

#Dataframes from csv files.
df_calc_test = pd.read_csv('calc_case_description_test_set.csv')
df_calc_train = pd.read_csv('calc_case_description_train_set.csv')
df_mass_test = pd.read_csv('mass_case_description_test_set.csv')
df_mass_train = pd.read_csv('mass_case_description_train_set.csv')
```

```python
#Dataframe columns indexes.
print(df_calc_test.columns)
```

```python
#Check number of mammography ROI lesion in each dataset - 
#it is related to the selected columns.
columns = ['patient_id', 'left or right breast', 'image view', 'abnormality id']

num_cases_calc_test = df_calc_test[columns].drop_duplicates().shape[0]
num_cases_calc_train = df_calc_train[columns].drop_duplicates().shape[0]
num_cases_mass_test = df_mass_test[columns].drop_duplicates().shape[0]
num_cases_mass_train = df_mass_train[columns].drop_duplicates().shape[0]

print('Total number of ROI lesion images:')
print(num_cases_calc_test)
print(num_cases_calc_train)
print(num_cases_mass_test)
print(num_cases_mass_train)
```

```python
#Check number of mammography ROI lesion in each dataset classified as MALIGNANT - 
#it is related to the selected columns.
df_calc_test_MALIGNANT = df_calc_test [df_calc_test['pathology'] == 'MALIGNANT']
df_calc_train_MALIGNANT = df_calc_train [df_calc_train['pathology'] == 'MALIGNANT']
df_mass_test_MALIGNANT = df_mass_test [df_mass_test['pathology'] == 'MALIGNANT']
df_mass_train_MALIGNANT =df_mass_train [df_mass_train['pathology'] == 'MALIGNANT']
```

```python
num_cases_calc_test_MALIGNANT = df_calc_test_MALIGNANT[columns].drop_duplicates().shape[0]
num_cases_calc_train_MALIGNANT = df_calc_train_MALIGNANT[columns].drop_duplicates().shape[0]
num_cases_mass_test_MALIGNANT = df_mass_test_MALIGNANT[columns].drop_duplicates().shape[0]
num_cases_mass_train_MALIGNANT = df_mass_train_MALIGNANT[columns].drop_duplicates().shape[0]

print('Total number of ROI lesion images - MALIGNANT:')
print(num_cases_calc_test_MALIGNANT)
print(num_cases_calc_train_MALIGNANT)
print(num_cases_mass_test_MALIGNANT)
print(num_cases_mass_train_MALIGNANT)
```

```python
final_table = pd.DataFrame({
    'Dataset': [
        'Calcification Test',
        'Calcification Train',
        'Mass Test',
        'Mass Train'
    ],
    'Number of cases': [
        num_cases_calc_test,
        num_cases_calc_train,
        num_cases_mass_test,
        num_cases_mass_train
    ],
    'Benign': [
        #Difference between number total of images and the 
        #number of images classified as MALIGNANT.
        num_cases_calc_test - num_cases_calc_test_MALIGNANT,
        num_cases_calc_train - num_cases_calc_train_MALIGNANT,
        num_cases_mass_test - num_cases_mass_test_MALIGNANT,
        num_cases_mass_train - num_cases_mass_train_MALIGNANT
    ],
    'Malignant': [
        num_cases_calc_test_MALIGNANT,
        num_cases_calc_train_MALIGNANT,
        num_cases_mass_test_MALIGNANT,
        num_cases_mass_train_MALIGNANT
    ]
})

print(final_table)
```
